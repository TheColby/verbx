# Corpus pipeline orchestration

```python
import subprocess
from pathlib import Path

Path('nb_corpus').mkdir(exist_ok=True)
cmd = [
    'python', '-m', 'verbx.cli', 'batch', 'corpus-generate', 'nb_in.wav',
    '--output-root', 'nb_corpus',
    '--variants-per-input', '8',
    '--jobs', '2',
    '--retries', '1',
]
subprocess.run(cmd, check=True)
print('Generated corpus in', Path('nb_corpus').resolve())
```
