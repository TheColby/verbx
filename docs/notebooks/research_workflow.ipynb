# verbx research notebook (minimal)

This notebook demonstrates the stable `verbx.api` surface for file analysis, IR synthesis, and rendering.

```python
from pathlib import Path
import json
from verbx.api import analyze_file, generate_ir, render_file
```

```python
infile = Path("../audio/dry_click.wav")
out_ir = Path("./tmp_research_ir.wav")
out_render = Path("./tmp_research_render.wav")
```

```python
ir_report = generate_ir(out_ir, mode="hybrid", length=1.0, sr=48000, channels=1, seed=7)
analysis = analyze_file(infile)
render_report = render_file(infile, out_render, engine="conv", ir=str(out_ir), wet=0.7, dry=0.3, progress=False, silent=True)
print(json.dumps({"ir": ir_report["outfile"], "rms": analysis["rms"], "engine": render_report["engine"]}, indent=2))
```
