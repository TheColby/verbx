# Render + Analyze with `verbx.api`

```python
import numpy as np
import soundfile as sf
from verbx.api import render_file, analyze_file

sr = 24000
x = np.zeros((2048, 1), dtype=np.float64)
x[100:300, 0] = 0.25
sf.write('nb_in.wav', x, sr)

render_file('nb_in.wav', 'nb_out.wav', engine='algo', rt60=0.5, wet=0.3, dry=0.9)
metrics = analyze_file('nb_out.wav', include_loudness=True)
metrics
```
