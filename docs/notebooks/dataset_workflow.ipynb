# verbx dataset notebook (minimal)

Batch-style dataset analysis using `verbx.api.analyze_file`.

```python
from pathlib import Path
import json
from verbx.api import analyze_file
```

```python
dataset_dir = Path("../audio")
paths = sorted(dataset_dir.glob("*.wav"))[:5]
rows = []
for path in paths:
    metrics = analyze_file(path)
    rows.append({"file": path.name, "duration": metrics["duration"], "rms_dbfs": metrics["rms_dbfs"], "spectral_centroid": metrics["spectral_centroid"]})
print(json.dumps(rows, indent=2))
```
